```python
# how to use the numpy where
# First, slow way to do this
A=np.array([1,2,3,4])
B=np.array([100,200,300,400])
condition =np.array([True,True,False,False])  # Boolean Array
answer = [(A_val if cond else B_val) for A_val, B_val, cond in zip(A,B,condition)] # a list comprehension
answer
```

```python
# Next, use numpy.where
answer2= np.where(condition, A, B) # where condition is met, show A. Otherwise show B. Argument (cond, A,B)
answer2
```
