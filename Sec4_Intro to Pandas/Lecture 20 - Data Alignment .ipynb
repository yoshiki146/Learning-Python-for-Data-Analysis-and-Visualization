```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
ser1=Series([0,1,2], index=["A","B","C"] )
ser1
```

```python
ser2=Series([3,4,5,6], index=["A","B","C","D"])
ser2
```

```python
ser1+ser2 # D cannot computed
```

```python
dframe1=DataFrame(np.arange(4).reshape((2,2)),columns=list("AB"), index=["NYC","LA"]) # `list("AB")` returns "A" and "B"
dframe1
```

```python
dframe2=DataFrame(np.arange(9).reshape((3,3)), columns=list("ADC"),index=["NYC","SF","LA"])
dframe2
```

```python
dframe1+dframe2 # only the boxes that have inputs from both can be computed. Others, NaN
```

```python
dframe1.add(dframe2,fill_value=0) # if you wanna add diff size DF. Dont forget the argu fill_value=0
```

```python
ser3= dframe2.ix[0]
ser3
```

```python
dframe2+ser3   # adds [0,1,2] to all rows
```

```python
dframe2-ser3  # subtracts [0,1,2] from all rows
```
