```python
dframe=DataFrame([[1,2,3],[np.nan,5,6],[7,np.nan,9],[np.nan,np.nan,np.nan]])
dframe
```

```python
clean_dframe=dframe.dropna() # select only complete row 
clean_dframe
```

```python
dframe.dropna(how="all")  #  drop a row with all NA
```

```python
dframe.dropna(axis=1) # select only complete column
```

```python
npn=np.nan
dframe2=DataFrame([[1,2,3,npn],[2,npn,5,6],[npn,7,npn,9],[1,npn,npn,npn]])
dframe2
```

```python
dframe2.dropna(thresh=2)  # select a row with at least two data points, thresh:threshold
```

```python
dframe2.fillna(1)  # fill na with 1
```

```python
dframe2.fillna({0:0,1:1,2:2,3:3})  # fill 0th colum with 0, 1st with 1, and so on
```

```python
dframe2.fillna(0, inplace=True)  # make a change permanent 
dframe2
```
