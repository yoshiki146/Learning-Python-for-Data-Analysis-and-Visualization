```python
dframe2= DataFrame(np.arange(16).reshape(4,4),index=[['a','a','b','b'],[1,2,1,2]],     # `index` for rows. `[[higher index],[lower index]]`
                                       columns=[['NY','NY','LA','SF'],['cold','hot','hot','cold']])
dframe2
```

```python
dframe2.index.names= ["INDEX_1","INDEX_2"]
dframe2.columns.names=["Cities","Temp"]
dframe2
```

```python
# interchange the level
dframe2.swaplevel("Cities","Temp",axis=1)
```

```python
# sort level
dframe2.sortlevel(1)  # `1` means the second index. `axis=0` by default
```

```python
dframe2.sum(level="Temp",axis=1) # compute a sum on a certain level
```
