```python
ww2_cas = Series([8700000,4300000,3000000,2100000,400000], index=["USSR","Germany","China","Japan","USA"]) # casualty of ww2
ww2_cas
```

```python
ww2_cas['USA']
```

```python
# Check which countries had cas > 4 mil
ww2_cas[ww2_cas>4000000]
```

```python
# Check if an index is included in the series
"USSR" in ww2_cas
```

```python
# convert series into dictionary
ww2_dict = ww2_cas.to_dict()
ww2_dict
```

```python
ww2_series = Series(ww2_dict)  # from dict into series
ww2_series
```

```python
countries = ["China","Germany","Japan","USA","USSR","Argentina"]
```

```python
obj2= Series(ww2_dict,index=countries) # index is not necesarily input manually
obj2
```

```python
pd.isnull(obj2) # detect missing value
```

```python
pd.notnull(obj2)
```
