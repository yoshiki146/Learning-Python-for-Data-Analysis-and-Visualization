```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
dframe1=DataFrame(np.arange(8).reshape(2,4),
                                     index=pd.Index(["LA","SF"],name="city"),    # by using `pd.Index,` you can use `name="city"`
                                     columns=pd.Index(["A","B","C","D"],name="letter"))
dframe1
```

```python
# Use stack to pivot the columns into the rows
dframe_st=dframe1.stack()
dframe_st
```

```python
# rearrange back to DF
dframe_st.unstack()
```

```python
dframe_st.unstack("letter")
```

```python
dframe_st.unstack("city")
```

```python
ser1= Series([0,1,2],index=["Q","X","Y"])
ser2= Series([4,5,6],index=["X","Y","Z"])
```

```python
# concatenate Ser into DF
dframe= pd.concat([ser1,ser2],keys=["Alpha","Beta"])
dframe
```

```python
dframe.unstack()
```

```python
dframe.unstack().stack() #null values are eliminated if you stack
```

```python
dframe.unstack().stack(dropna=False) # to keep NA
```
