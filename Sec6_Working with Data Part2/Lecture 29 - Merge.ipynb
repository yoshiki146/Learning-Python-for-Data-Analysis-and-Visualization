```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
# create a DF through dictionary. Set up a column name and a list of data 
dframe1= DataFrame({"key":['X','Z','Y','Z','X','X'],"data_set_1":np.arange(6)}) 
dframe1
```

```python
dframe2= DataFrame({"key":['Q','Y','Z'],"data_set_2":[1,2,3]})
dframe2
```

```python
# merge dataframes
pd.merge(dframe1,dframe2)  # merge only complete data
```

```python
# specify which column to merge on
pd.merge(dframe1, dframe2, on="key")
```

```python
pd.merge(dframe1, dframe2, on="key", how="left") 
```

```python
pd.merge(dframe1,dframe2,on="key",how="right")  # merge the column that have the keys in the right DF
```

```python
pd.merge(dframe1, dframe2,on="key",how="outer") # union of both DFs
```

```python
dframe3= DataFrame({"key":['X', 'X', 'X', 'Y', 'Z', 'Z'],"data_set_3":range(6)})
dframe3
```

```python
dframe4= DataFrame({"key":["Y","Y","X","X","Z"],"data_set_4":range(5)})
dframe4
```

```python
pd.merge(dframe3,dframe4)
```

```python
df_left = DataFrame({'key1': ['SF', 'SF', 'LA'],
                  'key2': ['one', 'two', 'one'],
                  'left_data': [10,20,30]})
df_left
```

```python
#Dframe on right
df_right = DataFrame({'key1': ['SF', 'SF', 'LA', 'LA'],
                   'key2': ['one', 'one', 'one', 'two'],
                   'right_data': [40,50,60,70]})
df_right
```

```python
pd.merge(df_left,df_right,on=["key1","key2"],how="outer")
```

```python
# if you choose not to merge on the colums with the same name, pandas automatically adds suffix (key2_x, key2_y)
pd.merge(df_left, df_right, on="key1")
```

```python
# you can also specify suffices
pd.merge(df_left,df_right,on="key1",suffixes=("_lefty","_righty"))
```
