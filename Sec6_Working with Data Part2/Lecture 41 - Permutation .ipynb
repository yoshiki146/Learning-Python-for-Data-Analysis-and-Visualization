```python
box= np.array([1,2,3])
```

```python
shaker= np.random.randint(low=0,high=len(box),size=10) # creates random integers. note that high=3 isn't included [0,1,2,3)
shaker
```

```python
hand_grabs=box.take(shaker)   # grab the value from the indeces of shaker (shaker's indeces ->box's value)
hand_grabs
```
